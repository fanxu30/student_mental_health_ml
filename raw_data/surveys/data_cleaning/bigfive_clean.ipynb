```python
import pandas as pd
import numpy as np
```

```python
bigfive_df = pd.read_csv("https://raw.githubusercontent.com/fanxu30/student_mental_health_ml/refs/heads/main/surveys/BigFive.csv")
bigfive_df.head()
```

```python
import re

original_columns = bigfive_df.columns
cleaned_columns = [re.search(r'-\s*(.*)', col).group(1) if '-' in col else col for col in original_columns]
bigfive_df.columns = cleaned_columns

print(bigfive_df.columns.tolist())
```

```python
bigfive_df["1. Is talkative"].dropna().unique()
```

```python
# convert all the columns in ordinal variable except uid and type
ordinal_order = ['Disagree Strongly', 'Disagree a little', 'Neither agree nor disagree', 'Agree a little', 'Agree strongly']

for col in bigfive_df.columns:
    if col not in ['uid', 'type']:
        bigfive_df[col] = pd.Categorical(bigfive_df[col], categories=ordinal_order, ordered=True)
        bigfive_df[col] = bigfive_df[col].cat.codes + 1

bigfive_df.head()
```
