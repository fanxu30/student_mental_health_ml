```python
import pandas as pd
import numpy as np
```

```python
bigfive_df = pd.read_csv("https://raw.githubusercontent.com/fanxu30/student_mental_health_ml/refs/heads/main/raw_data/surveys/BigFive.csv")
bigfive_df.head()
```

```python
import re

original_columns = bigfive_df.columns
cleaned_columns = [re.search(r'-\s*(.*)', col).group(1) if '-' in col else col for col in original_columns]
bigfive_df.columns = cleaned_columns

print(bigfive_df.columns.tolist())
```

```python
bigfive_df["1. Is talkative"].dropna().unique()
```

```python
# convert all the columns in ordinal variable except uid and type
ordinal_order = ['Disagree Strongly', 'Disagree a little', 'Neither agree nor disagree', 'Agree a little', 'Agree strongly']

for col in bigfive_df.columns:
    if col not in ['uid', 'type']:
        bigfive_df[col] = pd.Categorical(bigfive_df[col], categories=ordinal_order, ordered=True)
        bigfive_df[col] = bigfive_df[col].cat.codes + 1

bigfive_df.head()
```

```python
bigfive_df.to_csv("/Users/Cindy_ggjx/student_mental_health_ml/student_mental_health_ml/cleaned_data/surveys/clean_bigfive.csv", index=False)
```

```python
PHQ_df = pd.read_csv("https://raw.githubusercontent.com/fanxu30/student_mental_health_ml/refs/heads/main/raw_data/surveys/PHQ-9.csv")
PHQ_df.head()
```

```python
PHQ_df["Little interest or pleasure in doing things"].dropna().unique()
```

```python
PHQ_df["Response"].dropna().unique()
```

```python
# convert all the columns in ordinal variable except uid and type and Response
PHQ_order = ['Not at all', 'More than half the days', 'Several days', 'Nearly every day']
PHQ_response_order = ['Not difficult at all', 'Somewhat difficult', 'Very difficult', 'Extremely difficult']

for col in PHQ_df.columns:
    if col not in ['uid', 'type','Response']:
        PHQ_df[col] = pd.Categorical(PHQ_df[col], categories=PHQ_order, ordered=True)
        PHQ_df[col] = PHQ_df[col].cat.codes + 1
    if col in ['Response']:
        PHQ_df[col] = pd.Categorical(PHQ_df[col], categories=PHQ_response_order, ordered=True)
        PHQ_df[col] = PHQ_df[col].cat.codes + 1

PHQ_df.head()
```
