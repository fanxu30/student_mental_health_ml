```python
import pandas as pd
import numpy as np
```

```python
bigfive_df = pd.read_csv("https://raw.githubusercontent.com/fanxu30/student_mental_health_ml/refs/heads/main/raw_data/surveys/BigFive.csv")
bigfive_df.head()
```

```python
import re

original_columns = bigfive_df.columns
cleaned_columns = [re.search(r'-\s*(.*)', col).group(1) if '-' in col else col for col in original_columns]
bigfive_df.columns = cleaned_columns

print(bigfive_df.columns.tolist())
```

```python
bigfive_df["1. Is talkative"].dropna().unique()
```

```python
# convert all the columns in ordinal variable except uid and type
ordinal_order = ['Disagree Strongly', 'Disagree a little', 'Neither agree nor disagree', 'Agree a little', 'Agree strongly']

for col in bigfive_df.columns:
    if col not in ['uid', 'type']:
        bigfive_df[col] = pd.Categorical(bigfive_df[col], categories=ordinal_order, ordered=True)
        bigfive_df[col] = bigfive_df[col].cat.codes + 1

bigfive_df.head()
```

```python
bigfive_df.to_csv("/Users/Cindy_ggjx/student_mental_health_ml/cleaned_data/surveys/clean_bigfive.csv", index=False)
```

```python
PHQ_df = pd.read_csv("https://raw.githubusercontent.com/fanxu30/student_mental_health_ml/refs/heads/main/raw_data/surveys/PHQ-9.csv")
PHQ_df.head()
```

```python
PHQ_df["Little interest or pleasure in doing things"].dropna().unique()
```

```python
PHQ_df["Response"].dropna().unique()
```

```python
# convert all the columns in ordinal variable except uid and type and Response
PHQ_order = ['Not at all',  'Several days','More than half the days', 'Nearly every day']
PHQ_response_order = ['Not difficult at all', 'Somewhat difficult', 'Very difficult', 'Extremely difficult']

for col in PHQ_df.columns:
    if col not in ['uid', 'type','Response']:
        PHQ_df[col] = pd.Categorical(PHQ_df[col], categories=PHQ_order, ordered=True)
        PHQ_df[col] = PHQ_df[col].cat.codes
    if col in ['Response']:
        PHQ_df[col] = pd.Categorical(PHQ_df[col], categories=PHQ_response_order, ordered=True)
        PHQ_df[col] = PHQ_df[col].cat.codes
PHQ_df['total_score'] = PHQ_df.iloc[:, 2:-1].sum(axis=1)
bins = [0, 4, 9, 14, 19, 27]
labels = [
    'Minimal depression',
    'Mild depression',
    'Moderate depression',
    'Moderately severe depression',
    'Severe depression'
]

PHQ_df['depression_severity'] = pd.cut(PHQ_df['total_score'], bins=bins, labels=labels, right=True)

PHQ_df.head()
```

```python
PHQ_df.to_csv("/Users/Cindy_ggjx/student_mental_health_ml/cleaned_data/surveys/clean_PHQ-9.csv", index=False)
```

```python
vr_12_df["Compared to one year ago, how would you rate your physical health in general now?"].dropna().unique()
```

```python
vr_health_order = ['Fair', 'Good', 'Very good', 'Excellent']
vr_activity_order = ['No, not limited at all', 'Yes, limited a little', 'Yes, limited a lot']
vr_time_order = ['No, none of the time', 'Yes, a little of the time', 'Yes, some of the time', 'Yes, most of the time','Yes, all of the time']
vr_frequency_order = ['Not at all','A little bit','Moderately','Quite a bit', 'Extremely']
vr_time_order_2 = ['None of the time','A little of the time','Some of the time','A good bit of the time','Most of the time']
vr_strong_order = ['Much worse','Slightly worse','About the same','Slightly better','Much better']

for col in vr_12_df.columns:
    if col in ['In general, would you say your health is']:
        vr_12_df[col] = pd.Categorical(vr_12_df[col], categories=vr_health_order, ordered=True)
        vr_12_df[col] = vr_12_df[col].cat.codes + 1
    elif col in ['Moderate activities, such as moving a table, pushing a vacuum cleaner, bowling or playing golf?','Climbing several flights of stairs?']:
        vr_12_df[col] = pd.Categorical(vr_12_df[col], categories=vr_activity_order, ordered=True)
        vr_12_df[col] = vr_12_df[col].cat.codes + 1
    elif col in ['Accomplished less than you would like.','Accomplished less than you would like..1','Were limited in the kind of work or other activities.',"Didn't do work or other activities as carefully as usual."]:
        vr_12_df[col] = pd.Categorical(vr_12_df[col], categories=vr_time_order, ordered=True)
        vr_12_df[col] = vr_12_df[col].cat.codes + 1
    elif col in ['During the past 4 weeks, how much did pain interfere with your normal work (including both work outside the home and housework)?']:
        vr_12_df[col] = pd.Categorical(vr_12_df[col], categories=vr_frequency_order, ordered=True)
        vr_12_df[col] = vr_12_df[col].cat.codes + 1
    elif col in ['How much of the time during the past 4 weeks: Have you felt downhearted and blue?','During the past 4 weeks, how much of the time has your physical health or emotional problems interfered with your social activities (like visiting with friends, relatives, etc.)?','How much of the time during the past 4 weeks: Have you felt calm and peaceful?','How much of the time during the past 4 weeks: Did you have a lot of energy?']:
        vr_12_df[col] = pd.Categorical(vr_12_df[col], categories=vr_time_order_2, ordered=True)
        vr_12_df[col] = vr_12_df[col].cat.codes + 1
    elif col in ['Compared to one year ago, how would you rate your physical health in general now?','Compared to one year ago, how would you rate your emotional problems (such as feeling anxious, depressed or irritable) now?']:
        vr_12_df[col] = pd.Categorical(vr_12_df[col], categories=vr_strong_order, ordered=True)
        vr_12_df[col] = vr_12_df[col].cat.codes + 1

vr_12_df.head()
```

```python
vr_12_df.to_csv("/Users/Cindy_ggjx/student_mental_health_ml/cleaned_data/surveys/clean_vr_12.csv", index=False)
```

```python
psqi_df = pd.read_csv("https://raw.githubusercontent.com/fanxu30/student_mental_health_ml/refs/heads/main/raw_data/surveys/psqi.csv")
psqi_df.head()
```

```python
psqi_df.columns = psqi_df.columns.str.strip()

# define some keywords to identify the target columns
psqi_df = psqi_df.drop(columns=[
    'During the past month, what time have you usually gone to bed at night?',
    'During the past month, how long (in minutes) has it usually taken you to fall asleep each night?',
    'When have you usually gotten up in the morning?',
    'During the past month, how many hours of actual sleep did you get at night? (This may be different than the number of hours you spent in bed.)'
])
psqi_df.head()
```

```python
psqi_df["a. Cannot get to sleep within 30 minutes"].dropna().unique()
```

```python
psqi_df["During the past month, how would you rate your sleep quality overall?"].dropna().unique()
```

```python
psqi_order = ['Not during the past month', 'Less than once week', 'Once or a twice week', 'Three or a more times week']
psqi_order_2 = ['Very bad','Fairly bad','Fairly good','Very good']
for col in psqi_df.columns:
    if col not in [
        'uid','type',"During the past month, how would you rate your sleep quality overall?"
        ]:
        psqi_df[col] = pd.Categorical(psqi_df[col], categories=psqi_order, ordered=True)
        psqi_df[col] = psqi_df[col].cat.codes + 1
    elif col in ["During the past month, how would you rate your sleep quality overall?"]:
        psqi_df[col] = pd.Categorical(psqi_df[col], categories=psqi_order_2, ordered=True)
        psqi_df[col] = psqi_df[col].cat.codes + 1

psqi_df.head()
```
